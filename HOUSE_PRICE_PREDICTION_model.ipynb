```python
X=house.drop('price', axis=1)
y=house['price']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split( X,y, test_size=.2, random_state=1)
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.linear_model import LinearRegression # imports the LinearRegression class
model = LinearRegression() # creates an instance of the LinearRegression class and assigns it to the variable model
model.fit(X_train,y_train) # calls the fit method on the model instance
```

```python
model.score(X_test,y_test)*100
```

```python
y_pred=model.predict(X_test)
y_pred
```

```python
model.coef_
```

```python
model.intercept_
```

```python
m=model.coef_
C=model.intercept_
```

```python
X_test
```

```python
def prediction (X_test):
  y_pred=[]
  for j in range(len (X_test)):
    x=X_test.iloc[j].values
    yp=sum(x*m)
    yp=yp+C
    y_pred.append(yp)
    return y_pred
X_test.iloc[0].values
```
